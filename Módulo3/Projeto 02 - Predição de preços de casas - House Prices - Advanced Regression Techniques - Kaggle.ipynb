# 7. Modelagem (1.0 ponto)

Divida os dados em teste e treino, construa o seu modelo e teste alterar os parâmetros utilizados por ele

```python
dados_limpos1.info()
```

```python
#SEPARAÇÃO X E Y

X = dados_limpos1.drop(["SalePrice"], axis = 1)
y = np.array(dados_limpos1["SalePrice"], dtype = 'str')

X_std = StandardScaler().fit_transform(X)
```

```python
#REGRESSÃO LINEAR
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size = 0.2, random_state = 10)
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

print('R2: %2f' % r2_score(y_test, y_pred))
print(' MAE: %2f' % mean_absolute_error(y_test, y_pred))
print('RMSE: %2f' % np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
#RANDOM FOREST REGRESSOR

X_train2, X_test2, y_train2, y_test2 = train_test_split(X_std, y, test_size = 0.2, random_state = 10)

# Modelo

rf = RandomForestRegressor(random_state=10)
rf.fit(X_train2, y_train2)
y_pred2 = rf.predict(X_test2)

print('R2: %2f' % r2_score(y_test2, y_pred2))
print(' MAE: %2f' % mean_absolute_error(y_test2, y_pred2))
print('RMSE: %2f' % np.sqrt(mean_squared_error(y_test2, y_pred2)))
```

```python
#RIDGE

X_train3, X_test3, y_train3, y_test3 = train_test_split(X_std, y, test_size = 0.2, random_state = 10)

# Modelo

ridge = Ridge(random_state=10)
ridge.fit(X_train3, y_train3)
y_pred3 = ridge.predict(X_test3)

print('R2: %2f' % r2_score(y_test3, y_pred3))
print(' MAE: %2f' % mean_absolute_error(y_test3, y_pred3))
print('RMSE: %2f' % np.sqrt(mean_squared_error(y_test3, y_pred3)))
```
